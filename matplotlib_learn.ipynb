```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = np.arange(10)
plt.plot(data)
```

```python
fig = plt.figure()#创建一个空窗口（画板）
ax1 = fig.add_subplot(2,2,1)#设置画布子图区，图像应该是2×2的（即最多4张图）
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)#设置且画出编号1到4的图
ax4 = fig.add_subplot(2,2,4)
plt.plot(np.random.randn(50).cumsum(), 'k--')#"k--"是一个线型选项，用于告诉matplotlib绘制黑色虚线图 ,如果没指定就画在最后一个子图上
ax1.hist(np.random.randn(100), bins=20, alpha=1)    #alpha表示颜色深浅，默认是蓝色
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30), color='g')
```

```python
fig1, axes1 = plt.subplots(2, 3)#matplotlib有一个更为方便的方法plt.subplots，它可以创建一个新的Figure，并将返还一个subplot对象数组给axes
axes1[0,1]
fig1.subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=0, hspace=1)
#wspace和hspace用于控制宽度和高度的百分比，可以用作subplot之间的间距
```

```python
fig2,axes2 = plt.subplots(2, 2, sharex = True, sharey = True) #x 和y轴保持同刻度
for i in range(2):
    for j in range(2):
        axes2[i,j].hist(np.random.randn(1000), bins = 50,color='k', alpha=0.6)# 绘制树状图
plt.subplots_adjust(wspace=0.3, hspace=0.3)#fig2.subplots_adjust也行
```

```python
plt.plot(np.random.randn(30).cumsum(),'go--')#color='g',marker ='o',linestyle='--' 
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')#plot图例，指明每条使用plt.legend的线
```

```python
fig3,ax = plt.subplots(1,1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
#调用set_xticks和set_xticklabels，一个是X轴刻度范围，一个是加标签，都生产对象，
#如果不存储给另一个对象就会打印除生成对象地址
abels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],rotation=0, fontsize='small')#rotation是倾斜角度，fontsize是字体大小
title = ax.set_title('My first matplotlib plot')
xlabel = ax.set_xlabel('Stages')
ylabel = ax.set_ylabel('ke')#   也可以用集合的方式
```

# 添加图例

```python
fig4,ax4 = plt.subplots(1,1)
ax4.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax4.plot(np.random.randn(1000).cumsum(), 'g--', label='two')
ax4.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
lengend = plt.legend(loc = 'best')#调用ax4.legend()或plt.legend()来自动创建图例,legend方法有几个其它的loc位置参数选项。请查看文档字符串（使用ax.legend?)
```

```python
fig5,ax5 = plt.subplots(1,1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)

ax5.add_patch(rect)
ax5.add_patch(circ)
ax5.add_patch(pgon)
ax5
#plt.savefig('figpath.png', dpi=400, bbox_inches='tight')  将图片保存下来
```

# 使用pandas和seaborn绘图

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),columns=['A', 'B', 'C', 'D'],index=np.arange(0, 100, 10))
df.plot()
```

```python
fig,axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(10),index = list('abcdefghij'))
data.plot.bar(ax = axes[0],color = 'g',alpha = 0.6)  #绘制水平柱状图
data.plot.barh(ax = axes[1],color = 'k',alpha = 1)    #绘制垂直柱状图
```

```python
df = pd.DataFrame(np.random.rand(6, 4),index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df.plot.bar(stacked=True, alpha=0.5)   #设置stacked = True,每行的值就会堆叠在一起
```
